# Data Cleaning

## opendata_project table

```python
opendata_project = pd.read_csv('opendata_project.csv')
print(opendata_project.shape) # shown number of (row, column)
opendata_project.head(3)
```

### Select columns

```python
opendata_project.columns
```

```python
opendata_project = opendata_project[['project_id',  'name_en',  'propertytype_id',
       'propertytype_name_en', 'price_min', 'latitude',
       'longitude', 'neighborhood_id', 'neighborhood_name_en',
        'subdistrict_id', 'subdistrict_name_en',
       'subdistrict_name_th', 'district_id', 'district_name_en',
       'district_name_th', 'province_id', 'province_name_en',
       'province_name_th', 'zipcode', 'count_elevator',
       'count_elevator_service', 'count_floor', 'count_parking', 
       'count_unit', 'count_unittype', 'facility_clubhouse',
       'facility_fitness', 'facility_meeting', 'facility_park',
       'facility_playground', 'facility_pool', 'facility_security']]
```

```python
opendata_project.head()
```

###Missing/Null Checking

```python
opendata_project.isnull().any() #false ไม่มีค่าว่าง true มีค่าว่าง
```

```python
opendata_project.dtypes #price_min ควรเป็นค่าตัวเลข แต่เป็น object
```

### dropna

```python
# ' price_min ' ราคาต่ำสุดโครงการ
print(len(opendata_project[opendata_project['price_min'].isnull()])) #we used dropna()
opendata_project = opendata_project.dropna(subset=['price_min'])
```

```python
# ' longitude '
opendata_project[opendata_project['longitude'].isnull()]
```

```python
opendata_project = opendata_project.dropna(subset=['longitude'])
```

```python
# 'subdistrict_id' รหัสตำบล
print(len(opendata_project[opendata_project['subdistrict_id'].isnull()])) 
opendata_project = opendata_project.dropna(subset=['subdistrict_id'])
```

```python
# 'zipcode' รหัสไปรษณีย์
print(len(opendata_project[opendata_project['zipcode'].isnull()])) 
opendata_project = opendata_project.dropna(subset=['zipcode'])
```

###fill Null with real world Knownlege

```python
#fill NA with Text
opendata_project['neighborhood_id'] = opendata_project['neighborhood_id'].fillna('Unknown') #รหัสย่าน 
opendata_project['neighborhood_name_en'] = opendata_project['neighborhood_name_en'].fillna('Unknown') 
```

```python
# จำนวนลิฟต์และชั้น
opendata_project[['count_floor', 'count_unit','count_unittype']]  = opendata_project[['count_floor', 'count_unit','count_unittype']].fillna(1) #จำนวนชั้นของตัวโครงการ, จำนวนยูนิตในโครงการ, จำนวนแบบ้านในโครงการ
opendata_project[['count_elevator','count_floor','count_parking', 'count_elevator_service']] = opendata_project[['count_elevator','count_floor','count_parking', 'count_elevator_service']].fillna(0) 
# สิ่งอำนวยความสะดวก
opendata_project[['facility_clubhouse','facility_fitness','facility_meeting','facility_park','facility_playground','facility_pool','facility_security']] = opendata_project[['facility_clubhouse','facility_fitness','facility_meeting','facility_park','facility_playground','facility_pool','facility_security']].fillna(0) 
```

```python
opendata_project.isnull().any()
```

```python
#save data
opendata_project.to_csv('opendata_project_clean.csv', index=False)
```

#Merge tables

```python
import pandas as pd
opendata_project_clean = pd.read_csv('opendata_project_clean.csv')
opendata_unittype_clean = pd.read_csv('opendata_unittype_clean.csv') #ตารางหลัก
opendata_eating_score_clean = pd.read_csv('opendata_eating_score_clean.csv')
opendata_living_score_clean = pd.read_csv('opendata_living_score_clean.csv')
opendata_train_station = pd.read_csv('opendata_train_station_clean.csv')
opendata_train_line= pd.read_csv('opendata_train_line_clean.csv')
```

```python
#เปลี่ยนชื่อ columns ที่ซ้ำ
opendata_unittype_clean.rename(columns={'count_parking': 'count_parking_unittype', 'count_elevator': 'count_elevator_unittype'}, inplace=True)
opendata_unittype_clean.head(3)
```

```python
data_trains_clean = opendata_train_station[['line_typt','line_th','status','station_th','tam_namt','prov_namt']]
```

```python
data_trains_clean.rename(columns={'tam_namt': 'subdistrict_name_th','prov_namt':'province_name_th'}, inplace=True);
```

```python
data_trains_clean.columns
```

```python
data_trains_clean
```

```python
data_trains_clean.isnull().any()
```

```python
data_trainl_clean = opendata_train_line[['name', 'area',]]
```

```python
data_trainl_clean.rename(columns={'name': 'line_th'}, inplace=True);
```

```python
data_trainl_clean
```

```python
merge2 =  opendata_project_clean.merge(opendata_unittype_clean, left_on='project_id', right_on='project_id')
merge2.head()
```

```python
merge3 = merge2.merge(opendata_eating_score_clean)
merge3
```

```python
merge4  = merge3.merge(opendata_living_score_clean)
merge4
```

```python
merge4.isnull().any()
```

```python
merge5 = merge4.merge(data_trains_clean)
merge5
```

```python
merge5.isnull().any()
```

```python
merge5.columns
```

```python
merge6 = merge5.merge(data_trainl_clean)
merge6
```

```python
merge6.isnull().any()
```

```python
data = [{'prov':'Nonthaburi','val':12},
        {'prov':'Bangkok','val':20},
        {'prov':'Pathumthani','val':6}   
]
pddf = pd.DataFrame(data)
pddf
```

```python
from pandas._libs.lib import map_infer

fig2=Figure(width=550,height=350)
map=folium.Map(location=[13.8621, 100.5144],zoom_start=4.5)

## วาด geographical data ลงบนแผนที่
map.choropleth(
     geo_data=th_geojson,
     data=pddf,
     columns=['prov', 'val'],
     key_on='feature.properties.name',
     fill_color='Spectral', 
     fill_opacity=0.6, 
     line_opacity=1,
     legend_name='random number',
     smooth_factor=0)

fig2.add_child(map)
```
